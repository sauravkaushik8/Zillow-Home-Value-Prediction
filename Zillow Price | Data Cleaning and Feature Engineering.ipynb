```python
train.drop('transactiondate',inplace=True, axis = 1)
```

```python
def clean_data(train):
    
    
    train.airconditioningtypeid.fillna(-999,inplace=True)
    train.architecturalstyletypeid.fillna(-999,inplace=True)
    train.decktypeid.fillna(-999,inplace=True)
    train.heatingorsystemtypeid.fillna(-999,inplace=True)
    train.buildingclasstypeid.fillna(-999,inplace=True)

    train['airconditioningtypeid'] = train['airconditioningtypeid'].astype('category')
    train['architecturalstyletypeid'] = train['architecturalstyletypeid'].astype('category')
    train['decktypeid'] = train['decktypeid'].astype('category')
    train['heatingorsystemtypeid'] = train['heatingorsystemtypeid'].astype('category')
    train['buildingclasstypeid'] = train['buildingclasstypeid'].astype('category')

#    train.transactiondate = pd.to_datetime(train.transactiondate)

#    train["txn_year"] = train["transactiondate"].dt.year
#    train["txn_month"] = train["transactiondate"].dt.month
#    train["txn_day"] = train["transactiondate"].dt.day
#    train["txn_hour"] = train["transactiondate"].dt.hour
#    train["txn_weekday"] = train["transactiondate"].dt.weekday
#    train["txn_week"] = train["transactiondate"].dt.week
    
    train['pid_Counts'] = train.groupby(['parcelid'])['parcelid'].transform('count')

    train['nb_Counts'] = train.isnull().sum(axis = 1)
    
    missing_df_prop = train.isnull().sum()/train.shape[0]
    missing_df_prop_l = missing_df_prop.T.tolist()


    missing_weighted = train.isnull()*missing_df_prop_l

    train['nb_count_w'] = missing_weighted.sum(axis = 1)


    dt_type = train.dtypes.reset_index()
    dt_type.columns = ['predictor', 'type']
    cols_cat = dt_type.predictor[dt_type.type == 'category']
    cols_obj = dt_type.predictor[dt_type.type == 'object']
    
    

    oh = pd.get_dummies(train[cols_cat])

    train.drop(cols_cat, axis=1, inplace=True)

    train = pd.concat([train, oh], axis = 1)

    cols_obj = ['hashottuborspa', 'propertycountylandusecode', 'fireplaceflag', 'taxdelinquencyflag']

    oh = pd.get_dummies(train[cols_obj])

    train.drop(cols_obj, axis=1, inplace=True)

    train = pd.concat([train, oh], axis = 1)
    
    return train
```

```python
train = clean_data(train)
```
